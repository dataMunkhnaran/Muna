##PIECHART

```python
import numpy as np
cars = ['AUDI', 'BMW', 'FORD']
 
data = [23, 17, 35]
fig=plt.figure(figsize=(4,3))
plt.pie(data, labels=cars)
plt.show()
plt.savefig('fig.png')
```

HISTOGRAM- Тодорхой муж дахь тархалтын график

```python
fig, ax = plt.subplots(figsize =(10, 7))
ax.hist(df['Олгосон дүн'], bins = [0, 50000, 100000, 200000,300000, 400000, 500000, 800000])
plt.xlabel("Олгосон зээлийн хэмжээ ")
plt.ylabel("Зээл авсан хүний тоо")
plt.show()
```

```python
fig, ax = plt.subplots(figsize =(10, 7))
x = [50, 100, 150,200, 250, 300, 350,400,450,500]
ax.hist( df['Ашигласан хоног'], x )
plt.ylabel("Ашигласан хүний тоо")
plt.xlabel("Ашигласан хоног")
plt.show()
```
